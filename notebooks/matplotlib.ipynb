```python
import numpy as np
import matplotlib.pyplot as plt
import random

x = np.random.random(50) * 100
y = np.random.random(50) * 100

x = ["Python", "HTML/CSS", "JS", "C", "Go"]
y = [25, 20, 40, 15, 5]
plt.scatter(x,y, c='#007bff', marker="*", s=150, alpha=0.3)# C is to change the colors of the points, marker is to change the shape of the points, s is to change the size of the markers, alpha is to alter the transparency which is good for data with loads of data in order to see where it's most point saturated

years = [2016 + x for x in range(11)]
weights = [88,81,83,82,89,92,82,75,80,82,87]
plt.plot(years, weights, c="g", lw=3, linestyle="--")
plt.bar(x, y, align="edge", width=0.5, edgecolor="green", lw=6) # Align=edge to take the label of the barchart to the starting edge of the bar, edgecolor is to modify the color of the bar's edge, width - you know that one

ages = np.random.normal(20, 1.5, 1000)
plt.hist(ages, bins=3, cumulative=True) # Bin used to split the range of my data into 3 equal-sized intervals, cumulative=true is used to cummulatively stack the data
explodes = [.1, 0, 0, 0, 0]
plt.pie(y, labels=x, explode=explodes, autopct="%.2f%%", pctdistance=.6, startangle=90) # explodes is to specify how much you'd like each slice to be pulled out from the circle, autopct is to specify the value of the slice in percentage, pctdistance is to specify the distance of the autopct value of each slice from the center of the circle, startangle is to specify the starting position of the piechart

heights = np.random.normal(172, 8, 300)
# plt.boxplot(heights)

#Plot customization
years = [2019 + x for x in range(8)]
income = [62 + random.randint(0,8) for _ in range(8)]
income_ticks = list(range(60,70))
plt.title("Income of John over the years (In USD)", fontsize=20, fontname="poppins") # To set the title as well as customize the font size and name
plt.xlabel("Years") # xlabel 
plt.ylabel("Income in dollars") # ylabel
plt.yticks(income_ticks, [f"${x}k" for x in income_ticks]) # ticks for typically the y axis when you wanna modify its values
plt.plot(years, income)

stock_a = [100 + random.randint(-2,2) for _ in range(7)]
stock_b = [101 + random.randint(-10,8) for _ in range(7)]
stock_c = [100 + random.randint(-5,15) for _ in range(7)]

plt.plot(stock_a, label="Tesla") # For plotting multiple graphs, and also with the legend of which you'd be needing to add a label that will be shown after you've called the plt.legend
plt.plot(stock_b, label="Microsoft")
plt.plot(stock_c, label="Nike")
plt.legend(loc="upper center") # The function that shows the legend. loc for specifing the location of the shown label

votes = [12, 7, 5, 9, 22]
people = ["A", "B", "C", "D", "E"]
plt.pie(votes)
plt.legend(labels=people) # Same label talk here but for pie chart

x1, y1 = (np.random.random(100) * 100).round(), np.random.random(100)
x2, y2 = np.arange(100), np.random.random(100)

plt.figure(1)# Figure is used to number the different plots you wanna do. The oga of multiple plots
plt.scatter(x1, y1)

plt.figure(2)# 2nd plot
plt.plot(x2, y2)

x = np.arange(100)
fig, axs = plt.subplots(2,2) # Another method for plotting various named plots. 2,2 means 4 different plot ie (0,0), (0,1), (1,0), (1,1)
axs[0,0].plot(x, np.sin(x))
axs[0,0].set_title("Sine wave")

axs[0,1].plot(x, np.cos(x))
axs[0,1].set_title("Cosine wave")

axs[1,0].plot(x, np.random.random(100))
axs[1,0].set_title("Random wave")

axs[1,1].plot(x, np.log(x))
axs[1,1].set_title("Logarithmic wave")

fig.suptitle("The 4 plots")
plt.tight_layout() # To fix overlaps by evenly and nicely spacing the different plots
plt.savefig("fourplots.png", dpi=1000) # .savefig is used to save your outputted plot. dpi specifies the quality of the soon-to-be saved image, the higher the better the quality
```

```python
#3D plotting and animations
ax = plt.axes(projection="3d")

x = np.random.random(100)
y = np.random.random(100)
z = np.random.random(100)
x = np.arange(0,50,0.1)
y = np.arange(0,50,0.1) # np.sin(x)
z = np.cos(x + y)

ax.scatter(x, y, z)
ax.set_title("3D plot")
ax.set_xlabel("xtest")
ax.set_ylabel("ytest")
ax.set_zlabel("ztest")
ax.plot(x, y, z)
ax.set_title("3D plot")
ax.set_xlabel("xtest")
ax.set_ylabel("ytest")
ax.set_zlabel("ztest")

# Surface plot
x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)

X, Y = np.meshgrid(x,y)
Z = np.sin(X) * np.cos(Y)
ax.plot_surface(X, Y, Z, cmap="Spectral")
ax.view_init(azim=0, elev=90) # For setting the initial position of the generated image

#Animation
heads_tails = [0,0]
for _ in range(100000):
    heads_tails[random.randint(0,1)] += 1
    plt.bar(["Heads", "Tails"], heads_tails, color=["red", "blue"])
    # plt.pause(0.001)
```
