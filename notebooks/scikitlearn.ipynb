```python
from sklearn import datasets # Getting our dataset for model training
from sklearn.model_selection import train_test_split # Function for splitting our data for training and testing
from sklearn.linear_model import LinearRegression as LR # Our learning algorithm(How the model is to learn). Your selected learning algorithm isn't always optimal for your type of data, so you need to go through a lot to find the best for your data
from sklearn.ensemble import HistGradientBoostingRegressor as GBR, RandomForestRegressor as RFR 
from sklearn.metrics import r2_score as r2 # Our evaluation metric to see how accurate our model's predicted y value is
from sklearn.preprocessing import PolynomialFeatures as PF # We do what it's called optimization to make our model smarter. One way to do that is by using polynomial features, i.e getting the existing features and expand them 
import joblib # The class for saving our model into a file

housing = datasets.fetch_california_housing()

x = housing.data
y = housing.target
poly = PF() #
x = poly.fit_transform(x) # polynomializing our data

train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.2, random_state=30)

model = LR() 
model.fit(train_x, train_y) # The part where our model actually does the learning and stores the "brain"


# for model in [LR(), GBR(), RFR(n_jobs=-1)]: # n_jobs is used to tell the pc the number of cores to use for precessing
#     model.fit(train_x, train_y)
#     print(f"model:{model}", r2(test_y,  model.predict(test_x)))


predicted_y = model.predict(test_x) # Model evaluation with our test data to make sure our model didn't just memorized but learned patterns from our data 

score = r2(test_y, predicted_y) # r2 score is basically a score to know how much of the data your model understand and not it's numerical guess accuracy(It getting ...% of values correctly)

#score # Base line score of how much the model understands the data

# After selecting the best algorithm, we then proceed to hyperparameterization, en gros, tweak tons of parameters to further boost the accuracy

# for j in [0.1, 0.075, 0.05]:
#     for i in [200, 250, 300, 350, 400]: # n_jobs is used to tell the pc the number of cores to use for precessing
#         model = GBR(max_iter=300, learning_rate=0.075).fit(train_x, train_y)
#         print(f"model max iter tweak at {i} and learning rate at {j}", r2(test_y,  model.predict(test_x)))

# After hyperparameterization(In reality, instead of just tweaking a few like the way we did 2, you'll have to do so with as many parameters as possible to get a lot closer to 100% accuracy) fix your values and now it's time to save the learned model
model = GBR(max_iter=300, learning_rate=0.075).fit(train_x, train_y)
joblib.dump(model, "scikit_learn_housing_prediction.joblib")
```
