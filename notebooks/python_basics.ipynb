```python
##PYTHON BASICS

#Variables and Data Types
name = "Alice"  # String
age = 30        # Integer
height = 5.5    # Float
is_student = False  # Boolean
iterable = 0#Anything that can be looped over, like lists, strings, tuples, dictionaries, sets, etc. range(5)  # Range object

#Lists and Dictionaries
#Strings are lists of characters so they can be indexed like lists, and list functions can be used on them.
fruits = ["apple", "banana", "cherry"]  # List
print(fruits[0])  # Accessing first element
print(fruits[:])  # Slicing the list
print(len(fruits))  # Length of the list
fruits.append("date")  # Adding an element

person = {"name": "Bob", "age": 25}  # Dictionary
print(person)

#Escape Characters are used to insert special characters in a string.
print("Hello\nWorld")  # New line
print("She said, \"Hello!\"")  # Double quotes inside string
print("C:\\Users\\Alice")  # Backslash
print("First Line\rSecond Line")  # Carriage return
print("Column1\tColumn2")  # Tab

#Formated Strings, used for printing variables within strings
print(f"{name} is {age} years old and {height} feet tall.")

#String Methods
sample_text = "  Hello, World!  "
print(sample_text.lower())  # Lowercase
print(sample_text.upper())  # Uppercase
print(sample_text.strip())  # Remove whitespace  from both ends
print(sample_text.replace("World", "Python"))  # Replace substring
print(sample_text.split(","))  # Split string into list
print(sample_text.find("World"))  # Find substring index
print(sample_text.startswith("  He"))  # Check starts with substring
print(sample_text.endswith("!  "))  # Check ends with substring

import math #Importing Math Module for math operations
#Number data types and Methods
int_num = 10
floatnum = 15.6789
complex_num = 2 + 3j


#Number Methods
print(round(floatnum))  # Round to nearest integer
print(abs(-int_num))  # Absolute value
print(pow(int_num, 2))  # Power
print(max(5, 10, 3))  # Maximum value
print(min(5, 10, 3))  # Minimum value
print(int_num + floatnum)  # Addition
print(floatnum - int_num)  # Subtraction
print(int_num * 2)  # Multiplication
print(floatnum / 3)  # Division
print(int_num % 3)  # Modulus
print(int_num // 3)  # Floor division
print(math.sqrt(16))  # Square root
print(math.ceil(4.2))  # Ceiling
print(math.floor(4.7))  # Floor

#Comparison operators are used to compare two values.
print(age == 30)  # Equal to
print(age != 25)  # Not equal to
print(age > 18)   # Greater than
print(age < 40)   # Less than
print(age >= 30)  # Greater than or equal to
print(age <= 35)  # Less than or equal to


#Conditional Statements are used to perform different actions based on different conditions.
#If, Elif, Else
if age > 18:
    print(f"{name} is senior citizen.")
elif age == 18:
    print(f"{name} is just an adult.")
else:
    print(f"{name} is a minor.")

#Ternary Operator
status = "Adult" if age >= 18 else "Minor"
print(status)

#Logical Operators are used to combine conditional statements.
is_adult = age >= 18
is_employed = True
if is_adult and is_employed:
    print(f"{name} is an employed adult.")

if is_adult or is_student:
    print(f"{name} is either an adult or a student.")

#Chain Comparison
if 18 <= age <= 65:
    print(f"{name} is of working age.")

```

```python
#Loops are used for iterating over a sequence (like a list, tuple, dictionary, set, or string).
#For Loop
for fruit in fruits:
    print(fruit)

for number in range(1, 6):
    print(number)  

#For else Loop
for number in range(1, 4):
    print(number)
    if(number == 2):
        break
else:
    print("Number didn't get to 2.") 

#Nested Loop
for i in range(1, 4):
    for j in range(1, 4):
        print(f"i: {i}, j: {j}")

#While Loop
count = 0
while count < 5:
    print(count)
    count += 1

#Loop Control Statements
for number in range(1, 10):
    if number % 2 == 0:
        continue  # Skip even numbers
    if number > 7:
        break  # Exit loop if number is greater than 7
    print(number)

#Infinite Loop (Commented out to prevent execution)
# while True:
#     print("This is an infinite loop.")
#     break  # Use break to exit the loop
```

```python
#Functions are defined using the def keyword, followed by the function name and parentheses. Parameters can be included within the parentheses.
def greet(person_name):
    """Function to greet a person."""
    return f"Hello, {person_name}!"

print(greet("Alice")) #Calling the function. The value "Alice" is passed to the parameter person_name, and it's called an argument.

#Keyword Arguments
def describe_person(name, age, city):
    """Function to describe a person."""
    return f"{name} is {age} years old and lives in {city}."
print(describe_person(age=30, city="New York", name="Bob"))  # Using keyword arguments

#Default Parameters
def power(base, exponent=2):
    """Function to calculate power with default exponent 2."""
    return base ** exponent
print(power(3))  # Uses default exponent
print(power(2, 3))  # Overrides default exponent

#*args
def sum_all(*args):#For variable number of arguments
    """Function to sum all arguments."""
    return sum(args)

#**args
def print_info(**kwargs):#For keyword variable number of arguments
    """Function to print key-value pairs."""
    for key, value in kwargs.items():
        print(f"{key}: {value}")

#Lambda Functions. Lambda functions are small anonymous functions defined with the lambda keyword that don't stand alone like regular functions, but can be used as an argument to higher-order functions or assigned to variables.
square = lambda x: x ** 2 
```

```python
#Lists are ordered, mutable (changeable), and allow duplicate elements.
my_list = [1, 2, 3, 4, 5]
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]  # 2D List (Matrix)
zeros = [0] * 5  # List of five zeros
combined = my_list + zeros  # Concatenation
my_list = list(range(1, 6))  # Creating a list from a range
my_list = list("hello")  # Creating a list from a string

#Accessing Elements
print(my_list[0])  # First element
print(my_list[-1])  # Last element
print(my_list[1:4])  # Slicing
print(my_list[1::4])  # Step slicing
print(matrix[1][2])  # Accessing element in 2D list
#Modifying Lists
my_list[2] = 10  # Change value at index 2
my_list.append(6)  # Add element to end
my_list.insert(0, 0)  # Insert element at index 0
my_list.remove(3)  # Remove first occurrence of value 3
popped_element = my_list.pop()  # Remove and return last element
my_list.sort(key=lambda x: x[1] or sort_key)  # Sort the list
sorted_list = sorted(my_list)  # Return a new sorted list
#Sorting a list using a reference function
def sort_key(x):
    return x[1]  # Sort by second element of tuple
my_list.reverse()  # Reverse the list
my_list.clear()  # Clear the list
my_list.extend([7, 8, 9])  # Extend list with another list
my_list.count(2)  # Count occurrences of value 2
my_list.index(4)  # Find index of value 4

#List unpacking
a, b, c, d, e = my_list  # Unpacking list into variables
a, *rest, last = my_list  # Unpacking with rest
print(a, b, c, d, e)

for item in enumerate(my_list):
    print(item)  # Prints index and value as tuple


#Map function
squared_numbers = list(map(lambda x: x ** 2, my_list))  #Square each number in the list
print(squared_numbers)

#Filter function
even_numbers = list(filter(lambda x: x % 2 == 0, my_list))  #Filter even numbers from the list
print(even_numbers)

#List Comprehensions
squared = [x ** 2 for x in my_list]  # Squaring each number
evens = [x for x in my_list if x % 2 == 0]  # Filtering even numbers
pairs = [(x, y) for x in my_list for y in my_list]  # Cartesian product
print(squared)

#Zip function
list1 = [1, 2, 3]
list2 = ['a', 'b', 'c']
zipped = list(zip(list1, list2))  # Zipping two lists together

```

```python
#Arrays(Use Lists by default, but for specific use cases like with speed and memory efficiency, arrays can be used). Arrays are similar to lists but can only hold elements of the same data type, making them more memory efficient for large datasets.
import array as arr
numbers = arr.array('i', [1, 2, 3, 4, 5])  # 'i' is the type code for integers
print(numbers[0])  # Accessing first element

#Sets. Sets are unordered collections of unique elements. They don't support indexing or slicing like lists or tuples.
my_set = {1, 2, 3, 4, 5}
my_set = set([1, 2, 3, 4, 5])  # Creating a set from a list
second_set = {4, 5, 6, 7, 8}

#Set Operations
union_set = my_set.union(second_set)  # Union
intersection_set = my_set.intersection(second_set)  # Intersection
difference_set = my_set.difference(second_set)  # Difference
symmetric_difference_set = my_set.symmetric_difference(second_set)  # Symmetric Difference
my_set.add(6)  # Add element
my_set.remove(2)  # Remove element
my_set.discard(10)  # Remove element if exists, no error if not
my_set.pop()  # Remove and return an arbitrary element
is_member = 3 in my_set  # Membership test
set_size = len(my_set)  # Size of the set  
print(my_set | second_set)  # Union using | operator
print(my_set & second_set)  # Intersection using & operator
print(my_set - second_set)  # Difference using - operator
print(my_set ^ second_set)  # Symmetric Difference using ^ operator

#Dictionaries. These are collections of key-value pairs, where each key is unique and is used to access its corresponding value.
my_dict = {"name": "Alice", "age": 30, "city": "New York"}
my_dict = dict(name="Alice", age=30, city="New York")  # Alternative way to create a dictionary
second_dict = {"country": "USA", "occupation": "Engineer"}
#Dictionary Operations
value = my_dict["name"]  # Accessing value by key
my_dict["age"] = 31  # Modifying value
my_dict["email"] = "djflasfdh@dfjds;ljf.com"  # Adding new key-value pair
del my_dict["city"]  # Deleting key-value pair
my_dict.update(second_dict)  # Merging dictionaries
keys = my_dict.keys()  # Getting all keys
values = my_dict.values()  # Getting all values
items = my_dict.items()  # Getting all key-value pairs
is_key_present = "name" in my_dict  # Checking if key exists
dict_size = len(my_dict)  # Size of the dictionary
for key, value in my_dict.items():
    print(f"{key}: {value}")  # Iterating through key-value pairs
#Dictionary Comprehensions
squared_dict = {x: x ** 2 for x in range(1, 6)}  # Creating a dictionary with squares of numbers
print(squared_dict)
my_dict.get("name", "Unknown")  # Safe access with default value

#Generators. Generators are a type of iterable, like lists or tuples, but they don't store their contents in memory. Instead, they generate items on-the-fly and are defined using functions with the yield statement.

values = (x ** 2 for x in range(10))  # Generator expression
for val in values:
    print(val)  # Prints squares of numbers from 0 to 9

#Unpacking operators. You can use the unpacking operators * and ** to unpack any iterables and dictionaries respectively.
list_a = [1, 2, 3]
list_b = [4, 5, 6]
combined_list = [*list_a, *list_b]  # Unpacking lists into a new list
print(combined_list)
dict_a = {'x': 1, 'y': 2}
dict_b = {'y': 3, 'z': 4}
```
