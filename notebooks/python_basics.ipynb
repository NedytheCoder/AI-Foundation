```python
##PYTHON BASICS

#Variables and Data Types
name = "Alice"  # String
age = 30        # Integer
height = 5.5    # Float
is_student = False  # Boolean
iterable = 0#Anything that can be looped over, like lists, strings, tuples, dictionaries, sets, etc. range(5)  # Range object

#Lists and Dictionaries
#Strings are lists of characters so they can be indexed like lists, and list functions can be used on them.
fruits = ["apple", "banana", "cherry"]  # List
print(fruits[0])  # Accessing first element
print(fruits[:])  # Slicing the list
print(len(fruits))  # Length of the list
fruits.append("date")  # Adding an element

person = {"name": "Bob", "age": 25}  # Dictionary
print(person)

#Escape Characters are used to insert special characters in a string.
print("Hello\nWorld")  # New line
print("She said, \"Hello!\"")  # Double quotes inside string
print("C:\\Users\\Alice")  # Backslash
print("First Line\rSecond Line")  # Carriage return
print("Column1\tColumn2")  # Tab

#Formated Strings, used for printing variables within strings
print(f"{name} is {age} years old and {height} feet tall.")

#String Methods
sample_text = "  Hello, World!  "
print(sample_text.lower())  # Lowercase
print(sample_text.upper())  # Uppercase
print(sample_text.strip())  # Remove whitespace  from both ends
print(sample_text.replace("World", "Python"))  # Replace substring
print(sample_text.split(","))  # Split string into list
print(sample_text.find("World"))  # Find substring index
print(sample_text.startswith("  He"))  # Check starts with substring
print(sample_text.endswith("!  "))  # Check ends with substring

import math #Importing Math Module for math operations
#Number data types and Methods
int_num = 10
floatnum = 15.6789
complex_num = 2 + 3j


#Number Methods
print(round(floatnum))  # Round to nearest integer
print(abs(-int_num))  # Absolute value
print(pow(int_num, 2))  # Power
print(max(5, 10, 3))  # Maximum value
print(min(5, 10, 3))  # Minimum value
print(int_num + floatnum)  # Addition
print(floatnum - int_num)  # Subtraction
print(int_num * 2)  # Multiplication
print(floatnum / 3)  # Division
print(int_num % 3)  # Modulus
print(int_num // 3)  # Floor division
print(math.sqrt(16))  # Square root
print(math.ceil(4.2))  # Ceiling
print(math.floor(4.7))  # Floor

#Comparison operators are used to compare two values.
print(age == 30)  # Equal to
print(age != 25)  # Not equal to
print(age > 18)   # Greater than
print(age < 40)   # Less than
print(age >= 30)  # Greater than or equal to
print(age <= 35)  # Less than or equal to


#Conditional Statements are used to perform different actions based on different conditions.
#If, Elif, Else
if age > 18:
    print(f"{name} is senior citizen.")
elif age == 18:
    print(f"{name} is just an adult.")
else:
    print(f"{name} is a minor.")

#Ternary Operator
status = "Adult" if age >= 18 else "Minor"
print(status)

#Logical Operators are used to combine conditional statements.
is_adult = age >= 18
is_employed = True
if is_adult and is_employed:
    print(f"{name} is an employed adult.")

if is_adult or is_student:
    print(f"{name} is either an adult or a student.")

#Chain Comparison
if 18 <= age <= 65:
    print(f"{name} is of working age.")

```

```python
#Loops are used for iterating over a sequence (like a list, tuple, dictionary, set, or string).
#For Loop
for fruit in fruits:
    print(fruit)

for number in range(1, 6):
    print(number)  

#For else Loop
for number in range(1, 4):
    print(number)
    if(number == 2):
        break
else:
    print("Number didn't get to 2.") 

#Nested Loop
for i in range(1, 4):
    for j in range(1, 4):
        print(f"i: {i}, j: {j}")

#While Loop
count = 0
while count < 5:
    print(count)
    count += 1

#Loop Control Statements
for number in range(1, 10):
    if number % 2 == 0:
        continue  # Skip even numbers
    if number > 7:
        break  # Exit loop if number is greater than 7
    print(number)

#Infinite Loop (Commented out to prevent execution)
# while True:
#     print("This is an infinite loop.")
#     break  # Use break to exit the loop
```

```python
#Functions are defined using the def keyword, followed by the function name and parentheses. Parameters can be included within the parentheses.
def greet(person_name):
    """Function to greet a person."""
    return f"Hello, {person_name}!"

print(greet("Alice")) #Calling the function. The value "Alice" is passed to the parameter person_name, and it's called an argument.

#Keyword Arguments
def describe_person(name, age, city):
    """Function to describe a person."""
    return f"{name} is {age} years old and lives in {city}."
print(describe_person(age=30, city="New York", name="Bob"))  # Using keyword arguments

#Default Parameters
def power(base, exponent=2):
    """Function to calculate power with default exponent 2."""
    return base ** exponent
print(power(3))  # Uses default exponent
print(power(2, 3))  # Overrides default exponent

#*args
def sum_all(*args):#For variable number of arguments
    """Function to sum all arguments."""
    return sum(args)

#**args
def print_info(**kwargs):#For keyword variable number of arguments
    """Function to print key-value pairs."""
    for key, value in kwargs.items():
        print(f"{key}: {value}")

#Lambda Functions. Lambda functions are small anonymous functions defined with the lambda keyword that don't stand alone like regular functions, but can be used as an argument to higher-order functions or assigned to variables.
square = lambda x: x ** 2 
```
