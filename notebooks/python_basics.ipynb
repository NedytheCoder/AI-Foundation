```python
#Classes. Classes are blueprints for creating objects. They encapsulate data for the object and methods to manipulate that data.
class Person:
    def __init__(self, name, age):#Constructor method to initialize instance variables
        self.name = name  # Instance variable
        self.age = age    # Instance variable

    def greet(self):#Method to greet the person
        return f"Hello, my name is {self.name} and I am {self.age} years old."
#Creating an object of the Person class
person1 = Person("Alice", 30)

#Class and Instance Attributes
class Dog:
    species = "Canis familiaris"  # Class attribute

    def __init__(self, name, age):
        self.name = name  # Instance attribute
        self.age = age    # Instance attribute

    def bark(self):
        return f"{self.name} says Woof!"
    
dog1 = Dog("Buddy", 3)

#Class and Instance Methods
class Circle:
    pi = 3.14159  # Class attribute

    def __init__(self, radius):
        self.radius = radius  # Instance attribute

    def area(self):  # Instance method
        return Circle.pi * (self.radius ** 2)

    @classmethod
    def from_diameter(cls, diameter):  # Class method. Done with a decorator above it
        radius = diameter / 2
        return cls(radius)

    @staticmethod
    def circumference(radius):  # Static method
        return 2 * Circle.pi * radius 
    
rad = Circle.from_diameter(10)
print(rad)

#Magic Methods. These are special methods that start and end with double underscores (__). They allow you to define how objects of your class behave with
#built-in python functions and operators.
class Vector:
    #Private attributes can be created by prefixing the attribute name with double underscores (__). This makes the attribute inaccessible from outside the class, enforcing encapsulation.

    def __init__(self, x, y):
        self.__x = x
        self.y = y

    def __add__(self, other):  # Overloading the + operator
        return Vector(self.__x + other.__x, self.y + other.y)

    def __str__(self):  # String representation of the object
        return f"Vector({self.__x}, {self.y})"

    # Magic method for equality comparison
    def __eq__(self, other):
         return self.__x == other.__x and self.y == other.y
v1 = Vector(2, 3)
v2 = Vector(4, 5)
v3 = Vector(2, 3)

#Instance Object Comparison
print(v1 == v3)  # Using default equality comparison (compares memory addresses), without using the __eq__ method

class Product:
    def __init__(self, name, price):
        self.name = name
        #self.price = self.setprice(price)
        self.price = price

    def __repr__(self):  # Official string representation
        return f"Product(name={self.name}, price={self.price})"
    
    #Properties can be created using the @property decorator to define getter methods, and the @<property_name>.setter decorator to define setter methods. This allows controlled access to private attributes. They sit infront of an attribute and allows you to define methods that get or set the value of that attribute.

    @property
    def price(self):
        return self.price
    
    @price.setter #You don't always need to have both getter and setter for a property. You can have just one of them.
    def price(self, new_price):
        if new_price < 0:
            raise ValueError("Price cannot be negative.")
        self.price = new_price

    # Method 2 of initiating a property. price = property(getprice, setprice)

#Inheritance. Inheritance allows a class (child class) to inherit attributes and methods from another class (parent class). This promotes code reusability and establishes a hierarchical relationship between classes.

#Avoid multiple inheritance where possible, as it can lead to complex and hard-to-maintain code. Prefer single inheritance, double or composition over multiple inheritance(3 or more).
class Animal:
    def speak(self):
        return "Animal sound"
    def eat(self):
        return "Eating"
    def sleep(self):
        return "Sleeping"
    
class Cat(Animal):  # Inheriting from Animal class
    def speak(self):  # Overriding the speak method
        return "Meow"
    
print(Cat().speak())  # Output: Meow
print(Cat().eat())    # Output: Eating

class Person:
    def greet(self):
        print("Person Greet")

class Employee():  # Inheriting from Person class
    def greet(self):
        super().greet()  # Calling the parent class method
        print("Employee Greet")

#class Manager(Employee, Person):  # Multi-level Inheritance

#Abstract base classes. Abstract base classes are classes that cannot be instantiated on their own and are meant to be subclassed. They can contain abstract methods that must be implemented by any subclass. They are defined using the abc module.
from abc import ABC, abstractmethod

class Stream(ABC):
    def __init__(self):
        self.opened = False

    def open(self):
        if self.opened:
            raise ValueError("Stream is already open")
        self.opened = True
        print("Stream opened")

    def close(self):
        if not self.opened:
            raise ValueError("Stream is already closed")
        self.opened = False
        print("Stream closed")

    @abstractmethod
    def read(self):
        pass

class FileStream(Stream):
        def read(self):
            return "Reading data from file stream"

stream = FileStream().read()  # This will raise an error because Stream is an abstract class and cannot be instantiated directly.
print(stream)


#Polymorphism. Polymorphism allows methods to do different things based on the object it is acting upon, even if they share the same name. This is often achieved through method overriding in inheritance.

class Bird:
    def speak(self):
        return "Chirp"
class Dog:
    def speak(self):
        return "Woof" 
    

#Extending built-in types. You can create custom classes that extend built-in types like list, dict, or str to add additional functionality.
class MyList(list): #A list that extends the built-in list type and now able to get elements at even indices.
    def get_even_indices(self):
        return [self[i] for i in range(len(self)) if i % 2 == 0]
    

#Data classes. Data classes are a way to automatically generate special methods like __init__(), __repr__(), and __eq__() for classes that primarily store data. They are defined using the dataclasses module.
#Named tuples. Named tuples are an extension of regular tuples that allow you to access elements by name instead of index. They are defined using the collections module.
from collections import namedtuple #They act as data classes but are immutable like tuples.
Point = namedtuple('Point', ['x', 'y'])
p1 = Point(x=10, y=20)
p2 = Point(x=10, y=20)
print(p1==p2)
```

```python
#Machine Learning basics
#Steps include: Import data, Clean data, Split the data into training and test sets, Create the model using an algorithm, Train the model, Make predictions, Evaluate and improve the model.

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import joblib

music_data = pd.read_csv('csv_files/music.csv')  # Import data
print(music_data.columns)
#print(music_data.head())  # Display first few rows
music_input = music_data.drop(columns=['genre'])  # Features
music_target = music_data['genre']  # Target variable

model = DecisionTreeClassifier()  # Create model
X_train, X_test, y_train, y_test = train_test_split(music_input, music_target, test_size=0.2)  # Split data
model.fit(X_train, y_train)  # Train model
predictions = model.predict(X_test)  # Make predictions
print(predictions)  # Display predictions
score = model.score(X_test, y_test)  # Evaluate model
print(score)
print(accuracy_score(y_test, predictions))  # Calculate accuracy

joblib.dump(model, 'music_recommender.joblib')  # Save the model to a file
loaded_model = joblib.load('music_recommender.joblib')  # Load the model from where it's saved for persistence

loaded_predictions = loaded_model.predict(X_test)  # Make predictions with the loaded model
print(loaded_predictions)  # Display predictions from loaded model
```
